```python
#scrolling down

sleep(2)

pre_count = 0

while True:
    element = driver.find_elements(By.CSS_SELECTOR, "#subjects > div.list > table > tbody > tr")
    
    result = element[-1]
    driver.execute_script('arguments[0].scrollIntoView(true);',result)
    
    sleep(2)

    #현재 접근한 요소의 갯수
    current_count = len(element)
    if pre_count == current_count:
        break
    #같지않다면
    pre_count = current_count
```

```python
html = driver.page_source 
soup = BeautifulSoup(html, 'html.parser')

trs = soup.select('#subjects > div.list > table > tbody > tr')

results =[]
```

```python
for tr in range(len(trs)):
    result = {}
    #go though every column
    tds = trs[tr].select('#subjects > div.list > table > tbody > tr > td')
    result['CoursePl']=(tds[0].text) #Course Plan
    result['Category']=(tds[1].text) #Category
    result['Year']=(tds[2].text) #Year
    result['CourseID']=(tds[3].text) #Course Code
    result['CourseName']=(tds[4].text) #Course Name
    result['Credit']=int((tds[5].text)) #Credit
    result['Duration']=int((tds[6].text)) #Duration
    result['Professor']=(tds[7].text) #Professor Name
    result['Schedule']=(tds[8].text) #Schedule
    result['CurrentStudents']=int((tds[10].text)) #CurrentStudents
    result['TotalStudents']=int((tds[11].text)) #TotalStudents
    result['Note']=(tds[12].text) #Note
    results.append(result)
```

```python
import csv
fields = ['CoursePl',
'Category',
'Year',
'CourseID',
'CourseName',
'Credit',
'Duration',
'Professor',
'Schedule',
'CurrentStudents',
'TotalStudents',
'Note']
with open("CourseCatalogue.csv", "w" , encoding = "utf-8") as f:
    write = csv.writer(f)
    write.writerow(fields)
    for i in results:
        write.writerow(i.values())
```
